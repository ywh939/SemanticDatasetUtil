```python
import numpy as np
import open3d as o3d

def read_pcd_with_four_parameters(file_path):
    with open(file_path, 'r', encoding='latin-1') as file:
        lines = file.readlines()
    
    header_end = 0
    for i, line in enumerate(lines):
        if line.startswith('DATA'):
            header_end = i + 1
            break
    
    # Read points
    points = []
    for line in lines[header_end:]:
        values = line.strip().split()
        if len(values) == 4:  # Check if the point has 4 parameters
            try:
                point = list(map(float, values))
                points.append(point)
            except ValueError:
                print(f"Skipping invalid line: {line.strip()}")
    
    # Convert to NumPy array
    points_np = np.array(points)
    return points_np

points_np = read_pcd_with_four_parameters(out_file_path)

# Print the shape of the points array and the first few points to verify
print(f": {points_np.shape}")
```
